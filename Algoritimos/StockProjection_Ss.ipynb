*Projeção de estoque para o período de 1 ano, baseado em estratégia mista, ponto de pedido (s), ritmo de pedido (t) e no limite superior de estoque (S)*

```python
import pandas as pd 
import numpy as np
```

```python
stock_projection = pd.DataFrame(columns = ['Dia','Estoque',
'Necessidade'])
stock = 30000
s1 = 25500
s2 = 15500
order = 0
t1 = 0
T = 365

for i in range(1,T):
    
    demand = 8 + 428*np.random.beta(1.34, 1.45)
    stock_projection = stock_projection.append({'Dia': i,'Estoque': stock,'Necessidade': demand}, ignore_index=True)
    stock -= demand
    if i == t1:
        t1 = 0
        stock += order
        order = 0
    if stock < s2 and order == 0:
        L = round(np.random.normal(55.8,14.1))
        t1 = i + L
        order = s1 - stock
    if stock < 0:
        stock = 0

stock_projection
output = stock_projection.to_excel('StockProjection_material_3_sS_otimo.xlsx')
```
